```python
from pycaret.datasets import get_data
dataSets = get_data('index')
```

```python
from pycaret.clustering import *
from sklearn.datasets import load_wine

wine = load_wine()
ipl = wine.data


kMeanParameters1=setup(data=ipl, verbose=False)

x=create_model('kmeans')

print("For Cluster = 3")
x1 = create_model('kmeans', num_clusters = 3)

print("For Cluster = 4")
x1 = create_model('kmeans', num_clusters = 4)

print("For Cluster = 5")
x1 = create_model('kmeans', num_clusters = 5)

print("For Cluster = 6")
x1 = create_model('kmeans', num_clusters = 6)
```

```python
plot_model(x, plot = 'elbow')
```
